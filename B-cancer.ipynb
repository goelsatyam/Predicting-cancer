```python
#this function split the data frame in two parts
def split(data,p):
    a=[]
    b=[]
    n=data.shape[0]
    
    for i in range(n):
        x=random.random()
        if x<=p:
            a.append(i)
        else:
            b.append(i)
    return data.iloc[a],data.iloc[b]        
```

```python
#x is test data and y is train data
x,y=split(data,0.3)
```

```python
print x.shape[0],y.shape[0]
```

```python
#Logistic Regression
def sigmoid(z):
    return 1.0/(1.0+np.exp(-z))

def hypothesis(x,coef):
    h=coef[0]
    for i in xrange(1,len(coef)):
        h+=coef[i]*x[i-1]
  
    return sigmoid(h)
    
def gradient_decent(x,y,coef):
    n1=x.shape[0]
    b=np.array([0.0 for i in xrange(len(coef))],dtype=np.float)
    b1=np.array([[0.0 for i in range(len(coef))] for j in range(len(coef))])
    
    for i in xrange(n1):
        h=hypothesis(x.iloc[i],coef)
       
        b[0]+=(h-float(y.iloc[i]))*(1.0/(float(n1)))
        c1=[[1.0]]
        
        for j in xrange(1,len(coef)):
            b[j]+=(1.0/float(n1))*(h-float(y.iloc[i]))*x.iloc[i,j-1]
            c1.append([x.iloc[i,j-1]])
            
        c1=np.array(c1,dtype=np.float)
        c=c1.reshape((1,len(coef)))
        c1=c1.dot(c)
        c1=c1*h*(1-h)*(1.0/(float(n1)))
        b1+=c1
     
    H=np.linalg.inv(b1)
    H=H.dot(b)
    coef-=H
    return coef        

def LogisticRegression(x,y,iterations):
    coef=np.array([0.0 for i in xrange(x.shape[1]+1)],dtype=np.float)
    
    for iteration in xrange(iterations):
        coef=gradient_decent(x,y,coef)
    return coef    
```

```python
#training the data
b=LogisticRegression(y.drop('diagnosis',axis=1),y['diagnosis'],20)
```

```python
#calculating accuracy on test data
pre=0
p=x.drop(['diagnosis'],axis=1)
for i in range(p.shape[0]): 
    pred = hypothesis(p.iloc[i],b)
    if pred>=0.5:
        o=1.0
    else:
        o=0.0
        
    if o==x.iloc[i]['diagnosis']:
        pre+=1
        
print 'Accuracy :',pre/float(p.shape[0])*100        
```
